```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv", sep=",")
df.head()
```

```python
# See the first 10 entries.
df.head(10)
```

```python
# Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.
df = df.drop(df.columns[[0, 1]], axis=1)
df.head()
```

```python
# Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.
names = df.groupby("Name").sum().sort_values("Count", ascending=False)
names[["Count"]].head(10)
```

```python
# How many different names exist in the dataset?
num_uniq_names = df["Name"].nunique()
num_uniq_names
```

```python
# What is the name with most occurrences?
names["Count"].head(1)
```

```python
# What is the standard deviation of count of names?
names_std = names["Count"].std()
names_std
```

```python
# Get a summary of the dataset with the mean, min, max, std and quartiles.
summary = df.describe()
summary
```
