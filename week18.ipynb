```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

```python
file = "D:/VSCODE/Projects/Datasets/week18.csv"
df = pd.read_csv(file)
df.head()
```

1. Delete the Unnamed 0 Column

```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

2. Plot total_bill column hisogram

```python
# plt.style.use("seaborn")
sns.set()
fig, ax = plt.subplots()
ax.hist(df["total_bill"])
ax.set_xlabel("Total Bill")
plt.show()
plt.clf()
```

3.    Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",)

plt.show()
plt.clf()
```

4.    Create one image with the relationship of total_bill, tip and size.

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",
            size="size",
            hue="size",
            alpha=0.9)

plt.show()
plt.clf()
```

```python
sns.pairplot(df)
plt.show()
plt.clf()
```

5.    Present the relationship between days and total_bill value

```python
sns.relplot(x="day",
            y="total_bill",
            data=df,
            kind="line"
            )

plt.show()
plt.clf()
```

```python
sns.catplot(x="day",
            y="total_bill",
            data=df,
            kind="bar"
            )

plt.show()
plt.clf()
```

```python
sns.catplot(x="day",
            y="total_bill",
            data=df,
            kind="bar",
            errorbar="sd"
            )

plt.show()
plt.clf()
```

6.    Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.relplot(x="tip", y="day", 
            data=df,
            kind="scatter", 
            hue="sex")

plt.show()
plt.clf()
```

```python
sns.stripplot(x="tip", y="day", 
            data=df, 
            hue="sex")

plt.show()
plt.clf()
```

7.    Create a box plot presenting the total_bill per day differentiation the time (Dinner or Lunch)

```python
sns.catplot(x="day", y="total_bill",
            data=df,
            kind="box",
            sym="", 
            col="time")

plt.show()
plt.clf()
```

```python
sns.catplot(x="day", y="total_bill",
            data=df,
            kind="box",
            sym="",
            hue="time"
            )

plt.show()
plt.clf()
```

8.    Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.displot(x="tip",
             data=df,
             col="time"
             )

plt.show()
plt.clf()
```

9.    Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker

```python
sns.relplot(x="total_bill", y="tip",
            data=df,
            kind="scatter",
            col="sex", 
            hue="smoker",
            hue_order=["Yes", "No"])

plt.show()
plt.clf()
```
